```
!pip install panda
from pyspark import SparkContext
from pyspark import SparkConf

sc.stop()
conf = SparkConf().setMaster("local").setAppName("locations")
sc = SparkContext(conf=conf)
loadfile = sc.textFile('/content/drive/My Drive/Gowalla_totalCheckins.txt')

loadfile = loadfile.filter(lambda line: line.split(" "))\
        .map(lambda line:line.split("\t")[4]) 
        .map(lambda line: (line,1)) \
        .reduceByKey(lambda a,b:a+b) 
sorted(loadfile.collect(),key=lambda x:x[1],reverse=True)






```

```

sc.stop()
conf = SparkConf().setMaster("local").setAppName("users")
sc = SparkContext(conf=conf)
loadfile = sc.textFile('/content/drive/My Drive/Gowalla_totalCheckins.txt')
loadfile = loadfile.filter(lambda line: line.split(" ")) \
        .map(lambda line:line.split("\t")[0]) \
        .map(lambda line: (line,1)) \
        .reduceByKey(lambda a,b:a+b) 
   
sorted(loadfile.collect(),key=lambda x:x[1],reverse=True)
```

```
sc.stop()

conf = SparkConf().setMaster("local").setAppName("time"")
sc = SparkContext(conf=conf)
loadfile = sc.textFile('/content/drive/My Drive/Gowalla_totalCheckins.txt')
loadfile = loadfile.filter(lambda line: line.split(" ")) \
        .map(lambda line:line.split("\t")[1][11:13]) \
        .map(lambda line: (str(line)+":00:00 - "+str(line)+":59:59",1)) \
        .reduceByKey(lambda a,b:a+b) 
        
sorted(loadfile.collect(),key=lambda x:x[1],reverse=True)
```
